```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.worldometers.info/coronavirus/countries-where-coronavirus-has-spread/'
```

```python
response = requests.get(url)
response
```

```python
covid = BeautifulSoup(response.content, 'html.parser')
print(covid)
```

```python
# Getting data from the covid object
# As there are multiple 'td' tags we can use select method and than use square bracket to get the respective 'td' tag data, and than use .text method to extract the text data from the 'td' tag.
d1=covid.select('td')[0].text
d2=covid.select('td')[1].text
d3=covid.select('td')[2].text
d4=covid.select('td')[3].text
d5=covid.select('td')[4].text
d6=covid.select('td')[5].text
d7=covid.select('td')[6].text
d8=covid.select('td')[7].text
# Printing the first 8 data entries from the 'td' tag
d1,d2,d3,d4,d5,d6,d7,d8
```

```python
# As all the data is in 'td' tag ,using find_all method to get all the 'td' tags
data_iterator = covid.find_all('td')
data_iterator
```

```python
len(data_iterator)
```

```python
data = []
```

```python
# Using for loop to itter through the data and using steps as 4, and appending the data in data list 
# Replacing the ',' in confirmed and deaths by '' (i.e removing he ',' from the numbers and converig the data type to int)
for i in range (0,len(data_iterator),4):
        country = covid.select('td')[i].text
        confirmed = covid.select('td')[i+1].text
        deaths = covid.select('td')[i+2].text
        continent = covid.select('td')[i+3].text
        data.append((
            country,
            int(confirmed.replace(',', '')),
            int(deaths.replace(',', '')),
            continent
        ))
```

```python
import pandas as pd
```

```python
data
```

```python
df = pd.DataFrame(data)
```

```python
df[['Country','Confirmed_Cases','Deaths','Continent']] = df[[0,1,2,3]]
```

```python
df.rename(columns = {(0,1,2,3):('Country','Confirmed_Cases','Deaths','Continent')},inplace=True)
df
```

```python
df.columns = ['Country','Confirmed_Cases','Deaths','Continent']
df
```

```python
df.to_csv('Covid_2019.csv',index=False)
```
